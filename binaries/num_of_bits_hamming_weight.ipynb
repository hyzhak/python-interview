## Hamming Weight
https://en.wikipedia.org/wiki/Hamming_weight

```python
def num_of_bits(x):
    """
    Give number of bits in number x
    Algorithm Hamming weight (https://en.wikipedia.org/wiki/Hamming_weight)
    """
    print(f'input:{x:08b}')
    ans = 0
    while x:
        # move last bit to right 
        # and because all other bits should stay the same position
        # we would get all bigs expect the last one
        x &= x - 1 
        print(f'left: {x:08b}')
        ans += 1
    return ans

for x in range(16):
    print(f'bits: {x:08b}={num_of_bits(x)}\n')
```
